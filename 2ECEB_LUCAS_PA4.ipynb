```python
import pandas as pd                  #import pandas to access DataFrame

board2 = pd.read_csv('board2.csv')   #read csv file into a DataFrame
board2
```

1. Create the following data frames based on the format provided:
Example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as **Visayas**

    a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
Instru = board2[                               
    (board2['Electronics'] > 70) &             #filter rows where Electronics grade is greater than 70
    (board2['Track'] == 'Instrumentation') &   #filter rows where track is Instrumentation
    (board2['Hometown'] == 'Luzon')            #filter rows where hometown is Luzon
][['Name', 'GEAS', 'Electronics']]             #select only 'Name', 'GEAS', and 'Electronics' columns

Instru
```

    b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female    

```python
board2['Average'] = board2[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)   #calculate the average of the four subjects and create new column 'Average'

board2
```

```python
Mindy = board2[
    (board2['Average'] >= 55) &                  #filter rows where average grade is greater than or equal to 55
    (board2['Hometown'] == 'Mindanao') &         #filter rows where hometown is Mindanao
    (board2['Gender'] == 'Female')               #filter rows where gender is Female
][['Name', 'Track', 'Electronics', 'Average']]   #select only 'Name', 'Track', 'Electronics', and 'Average' columns

Mindy
```

2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
import matplotlib.pyplot as plt                                                  #import matplotlib to access visualization plots for data

# Track vs Average
track_avg = board2.groupby('Track')['Average'].mean()                            #group by 'Track' and calculate mean of 'Average'
plt.figure(figsize=(5,5))                                                        #set figure size
plt.bar(track_avg.index, track_avg.values, color='Maroon')                       #create bar chart with tracks on x-axis and mean average grade on y-axis
plt.xlabel('College Track')                                                      #label x-axis
plt.ylabel('Average Grade')                                                      #label y-axis
plt.title('College Track vs Average Score')                                      #set title
```

```python
# Gender vs Average
gender_avg = board2.groupby('Gender')['Average'].mean()                          #group by 'Gender' and calculate mean of 'Average'
plt.figure(figsize=(5,5))                                                        #set figure size
plt.bar(gender_avg.index, gender_avg.values, color='Green')                      #create bar chart with gender on x-axis and mean average grade on y-axis
plt.xlabel('Gender')                                                             #label x-axis
plt.ylabel('Average Grade')                                                      #label y-axis
plt.title('Gender vs Average Score')                                             #set title
```

```python
# Hometown vs Average
hometown_avg = board2.groupby('Hometown')['Average'].mean()                      #group by 'Hometown' and calculate mean of 'Average'
plt.figure(figsize=(5,5))                                                        #set figure size
plt.bar(hometown_avg.index, hometown_avg.values, color='Navy')                   #create bar chart with hometown on x-axis and mean average grade on y-axis
plt.xlabel('Hometown')                                                           #label x-axis
plt.ylabel('Average Grade')                                                      #label y-axis
plt.title('Hometown vs Average Score')                                           #set title
```
